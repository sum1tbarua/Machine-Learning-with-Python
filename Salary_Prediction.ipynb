# Salary Prediction using Linear Regression

```python
import pandas as pd 
import numpy as np
from matplotlib import pyplot as plt 
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
```

```python
# imported the csv file
df = pd.read_csv("Salary Data.csv", sep=',')

# Displaying first 5 records from the dataframe
df.head()
```

# EDA

```python
# independent variable/ feature variable
x = np.array(df['Experience Years']).reshape(-1, 1)
# dependent variable/ target variable
y = np.array(df['Salary'])

# plotting the values
plt.title('Salary Prediction')
plt.xlabel('Experience Years')
plt.ylabel('Salary')
plt.scatter(x, y)
plt.plot(x, y)
plt.show()
```

# Linear Regression

```python
# creating a linear regression class object
model = LinearRegression()

# Simple Linear Regression Equation: y = mx + c where m = coefficient and c = y-intercept

# Training the model
model.fit(x, y)

# Calculating the coefficient/slope
coefficient = model.coef_[0]

# Calculating the intercept
intercept = model.intercept_

# Calculating the model performence using R-Square value

# Predicting the y_values
y_predict = model.predict(x)

r2_value = r2_score(y, y_predict)

# We can also calculate the model performence using score method from LinearRegression class
score = model.score(x, y)

print(f"Coefficient: {coefficient:.3f}")
print(f"Intercept: {intercept:.3f}")
print(f"R2 Value: {r2_value:.3f}")
print(f"Is R2 Value and Score Equal? {score==r2_value}")

```

# Cost & Loss

```python
# Calculating the loss using Median Square Error

MSE = mean_squared_error(y, y_predict)
print(f"Mean Squared Error: {MSE:.3f}")
```
